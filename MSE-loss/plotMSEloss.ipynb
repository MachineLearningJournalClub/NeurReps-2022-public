```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

baseline = pickle.load( open("baselineModelBatchTraining.pkl", "rb") )

testLoss   =  baseline['testLoss'] * 9 / 8
trainLoss  =  baseline['trainLoss'] * 9

iterations = np.linspace(0, len(testLoss), len(testLoss))
fig, ax = plt.subplots(figsize = (10, 6))
ax.step(iterations, trainLoss, label="Training MSE loss", color="red")
ax.step(iterations, testLoss, label="Validation MSE loss", color="blue")
ax.grid(axis='x', color='0.85')
ax.grid(axis='y', color='0.85')
#ax.set_yscale('log')
ax.legend(title='MSE:')
plt.title('Loss function (MSE) vs number of iterations')
plt.xlabel('Iterations')
plt.ylabel('MSE value')
plt.savefig("lossPlotBaseline.pdf")
```
